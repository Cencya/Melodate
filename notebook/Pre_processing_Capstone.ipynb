### Adding User column for index

### Adding Religion column

```python
religions = ['islam', 'protestant christianity', 'roman catholicism', 'hinduism', 'buddhism', 'confucianism']

# Weights based on Indonesia's religious distribution 2023 (Wikipedia)
weights = [0.876, 0.0741, 0.0306, 0.0168, 0.0071, 0.0003]

# Assign religions based on the distribution and weights
general['religion'] = random.choices(religions, weights=weights, k=len(general))
```

# Merging both Dataset then Cleaning

```python
df_general = pd.read_csv('general_features.csv')
df_music = pd.read_csv('music_features.csv')
```

```python
# Choosing 300 random rows
df_general = df_general.sample(n=300, random_state=42)

# Reset index
df_general = df_general.reset_index(drop=True)

# Drop the 'religion' column
df_general = df_general.drop(columns=['religion'])
```

```python
religions = ['islam', 'protestant christianity', 'roman catholicism', 'hinduism', 'buddhism', 'confucianism']

df_general['religion'] = random.choices(religions, k=len(df_general))
```

```python
merged_df = pd.concat([df_general, df_music], axis=1)
```

```python
merged_df['user'] = ['user' + str(i) for i in range(len(merged_df))]
merged_df = merged_df[['user'] + [col for col in merged_df.columns if col != 'user']]
```

```python
merged_df.head(10)
```

```python
# Replace Only in specific situations (e.g., when working, studying) to "Only in specific situations"
merged_df['gender'] = merged_df['gender'].replace({'f': 'female', 'm': 'male'})

# Replace 'f' with 'female' and 'm' with 'male'
merged_df['listening_frequency'] = merged_df['listening_frequency'].replace({"Only in specific situations (e.g., when working, studying)": "Only in specific situations"})

# Replace married, seeing someone to 'in a relationship' and available to single
merged_df['status'] = merged_df['status'].replace({
    'married': 'in a relationship',
    'seeing someone': 'in a relationship',
    'available': 'single'
})
```
